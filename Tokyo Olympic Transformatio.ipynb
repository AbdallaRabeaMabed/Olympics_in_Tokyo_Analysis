```python
configs = {
  "fs.azure.account.auth.type.mabedtokyoanalysis1.dfs.core.windows.net": "OAuth",
  "fs.azure.account.oauth.provider.type.mabedtokyoanalysis1.dfs.core.windows.net": "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider",
  "fs.azure.account.oauth2.client.id.mabedtokyoanalysis1.dfs.core.windows.net": "",
  "fs.azure.account.oauth2.client.secret.mabedtokyoanalysis1.dfs.core.windows.net": "",
  "fs.azure.account.oauth2.client.endpoint.mabedtokyoanalysis1.dfs.core.windows.net": "https://login.microsoftonline.com/tanent_id/oauth2/token"
}

for key, value in configs.items(): spark.conf.set(key, value)
```

```python
#Loading the csv files from the ADSL After setting the configurations

athletes_df = spark.read.format("csv") \
  .option("header", "true") \
  .option("inferSchema", "true") \
  .load("abfss://tokyo-olympic-data@mabedtokyoanalysis1.dfs.core.windows.net/raw-data/athletes.csv")


coaches_df = spark.read.format("csv") \
  .option("header", "true") \
  .option("inferSchema", "true") \
  .load("abfss://tokyo-olympic-data@mabedtokyoanalysis1.dfs.core.windows.net/raw-data/coaches.csv")



entriesGender_df = spark.read.format("csv") \
  .option("header", "true") \
  .option("inferSchema", "true") \
  .load("abfss://tokyo-olympic-data@mabedtokyoanalysis1.dfs.core.windows.net/raw-data/entriesGender.csv")



medals_df = spark.read.format("csv") \
  .option("header", "true") \
  .option("inferSchema", "true") \
  .load("abfss://tokyo-olympic-data@mabedtokyoanalysis1.dfs.core.windows.net/raw-data/medals.csv")



teams_df = spark.read.format("csv") \
  .option("header", "true") \
  .option("inferSchema", "true") \
  .load("abfss://tokyo-olympic-data@mabedtokyoanalysis1.dfs.core.windows.net/raw-data/teams.csv")
```

```python
entriesGender_df = entriesGender_df.withColumn("Female", entriesGender_df["Female"].cast("int"))
entriesGender_df = entriesGender_df.withColumn("Male", entriesGender_df["Male"].cast("int"))
entriesGender_df = entriesGender_df.withColumn("Total", entriesGender_df["Total"].cast("int"))
```

```python
# Find the top countries with the highest number of gold medals
top_gold_medal_countries = medals_df.orderBy("Gold", ascending=False).select("TeamCountry","Gold").show()
```

```python
# Calculate the average number of entries by gender for each discipline
average_entries_by_gender = entriesGender_df.withColumn(
    'Avg_Female', entriesGender_df['Female'] / entriesGender_df['Total']
).withColumn(
    'Avg_Male', entriesGender_df['Male'] / entriesGender_df['Total']
)
average_entries_by_gender.show()
```

```python
athletes_df.write.format("csv") \
  .option("header", "true") \
  .mode("overwrite") \
  .save("abfss://tokyo-olympic-data@mabedtokyoanalysis1.dfs.core.windows.net/transformed-data/athletes")


coaches_df.write.format("csv") \
  .option("header", "true") \
  .mode("overwrite") \
  .save("abfss://tokyo-olympic-data@mabedtokyoanalysis1.dfs.core.windows.net/transformed-data/coaches")


entriesGender_df.write.format("csv") \
  .option("header", "true") \
  .mode("overwrite") \
.save("abfss://tokyo-olympic-data@mabedtokyoanalysis1.dfs.core.windows.net/transformed-data/entriesGender_df")

medals_df.write.format("csv") \
  .option("header", "true") \
  .mode("overwrite")    \
  .save("abfss://tokyo-olympic-data@mabedtokyoanalysis1.dfs.core.windows.net/transformed-data/medals")


teams_df.write.format("csv") \
  .option("header", "true") \
  .mode("overwrite")    \
  .save("abfss://tokyo-olympic-data@mabedtokyoanalysis1.dfs.core.windows.net/transformed-data/teams")

```
